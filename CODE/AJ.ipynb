```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data=pd.read_csv("data.csv")
data.columns=['time','North','East','West','South','NGR','EGR','WGR','SGR']
data.head()
```

```python
data['time']=data.index%96
data.head(100)
```

```python
sb.set(style='whitegrid',context='notebook')
cols=['time','North','East','West','South']
g=sb.pairplot(data[cols],size=3)
for i in range(0,4):
    g.axes[i,4].xaxis.set_label_text('time')
for i in range(4):
    g.axes[i,4].yaxis.set_label_text(cols[i])
plt.show()
```

```python
import numpy as np
cm = np.corrcoef(data[cols].values.T)
sb.set(font_scale=1.5)
hm=sb.heatmap(cm,
               cbar=True,
               annot=True,
               square=True,
               fmt='.2f',
               annot_kws={'size':15},
               yticklabels=cols,
               xticklabels=cols)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
from sklearn.cross_validation import train_test_split
X=data[cols].values
y=data[['NGR','EGR','WGR','SGR']]
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.25)
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_test, y_test)
```

```python
lr.predict([29,119,120,75,83])
```
